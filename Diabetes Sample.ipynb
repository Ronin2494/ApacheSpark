```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
from pyspark.sql.types import *
from pyspark.sql.functions import *

df = spark.read.load('Tables/diabetes')

#from a csv:  df = spark.read.load('Files/data/products.csv', format='csv', header=True, schema = productSchema)

display(df.limit(10))

#### to define a schema #####
# # productSchema = StructType([
#     StructField("ProductID", IntegerType()),
#     StructField("ProductName", StringType()),
#     StructField("Category", StringType()),
#     StructField("ListPrice", FloatType())
#     ])

AgeSex = df.select("AGE", "SEX").where((df['BMI'] == "24.0"))

display(AgeSex)

display(AgeSex.groupBy("SEX").count())

#save the dataframe in a parquet file
AgeSex.write.mode("overwrite").parquet('Files/AgeSex')

#partition df based on some column, say, Sex

df.write.partitionBy("SEX").mode('overwrite').parquet('Files/PartitionSex')

#load partitioned data
sexBased = spark.read.parquet('Files/PartitionSex/SEX=2')

display(sexBased.limit(5))


#create tempView
df.createOrReplaceTempView("products_view")



```
